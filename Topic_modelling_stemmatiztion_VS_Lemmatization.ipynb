```python
from nltk import corpus
import nltk
from nltk.tokenize import RegexpTokenizer
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import numpy as np 
from nltk.stem.porter import PorterStemmer
#from gensim import corpora
from gensim.corpora import Dictionary
```

 Lets look at the one of the text file of nltk corporus

```python
w_r=open("/home/preetham/nltk_data/corpora/gutenberg/austen-emma.txt","rt").read()
```

```python
w_r
```

```python
# converting paragraph into words

tokens = nltk.tokenize.word_tokenize(w_r)
# convert to lower case
tokens = [w.lower() for w in tokens]
# remove punctuation from each word
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
# remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]
# filter out stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
print(words)
```

```python
# stemming of words
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
print(stemmed)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print(lemmatized_words)
```

```python
print("Number of words:",len(words))
print("Number of words after Stemmatization:",len(stemmed))
print("Number of words after Lemmatization:",len(lemmatized_words))
```

```python
from collections import Counter

unique_stemmed=Counter(stemmed)
unique_lemma=Counter(lemmatized_words)
```

```python
print("Number of unique words after Stemmatization:",len(unique_stemmed))
print("Number of unique words after Lemmatization:",len(unique_lemma))
```
